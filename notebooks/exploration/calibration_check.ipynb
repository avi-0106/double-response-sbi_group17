```python
def check_double_rate(gap_scale, n_draws=10, n_trials=200):
    rates = []
    for _ in range(n_draws):
        p = prior()
        gap = np.random.exponential(scale=gap_scale)
        p['alpha2'] = p['alpha1'] + gap

        data = simulate_dataset(p['nu'], p['alpha1'], p['alpha2'], p['tau'], n=n_trials)
        rates.append(data['double_flag'].mean())

    print(f"gap_scale={gap_scale} -> rates: {[f'{r:.3f}' for r in rates]}")
    print(f"mean rate: {np.mean(rates):.3f}")

check_double_rate(gap_scale=5.0)
```

```python
def check_double_rate_v2(time_scale, n_draws=10, n_trials=200):
    rates = []
    for _ in range(n_draws):
        p = prior()
        nu = p['nu']
        alpha1 = p['alpha1']

        # how much EXTRA TIME (seconds) the loser gets, not raw distance
        extra_time = np.random.exponential(scale=time_scale)
        loser_speed = np.mean(nu)  # rough stand-in for whichever one loses
        gap = extra_time * loser_speed

        alpha2 = alpha1 + gap
        data = simulate_dataset(nu, alpha1, alpha2, p['tau'], n=n_trials)
        rates.append(data['double_flag'].mean())

    print(f"time_scale={time_scale} -> rates: {[f'{r:.3f}' for r in rates]}")
    print(f"mean rate: {np.mean(rates):.3f}")

check_double_rate_v2(time_scale=0.5)
check_double_rate_v2(time_scale=1.0)
check_double_rate_v2(time_scale=2.0)
```

```python
check_double_rate_v2(time_scale=5.0)
check_double_rate_v2(time_scale=10.0)
check_double_rate_v2(time_scale=20.0)
```

```python
def check_double_rate_v3(time_scale, time_shape, n_draws=10, n_trials=200):
    rates = []
    for _ in range(n_draws):
        p = prior()
        nu = p['nu']
        alpha1 = p['alpha1']

        extra_time = np.random.gamma(shape=time_shape, scale=time_scale / time_shape)
        loser_speed = np.mean(nu)
        gap = extra_time * loser_speed

        alpha2 = alpha1 + gap
        data = simulate_dataset(nu, alpha1, alpha2, p['tau'], n=n_trials)
        rates.append(data['double_flag'].mean())

    print(f"time_scale={time_scale}, shape={time_shape} -> rates: {[f'{r:.3f}' for r in rates]}")
    print(f"mean rate: {np.mean(rates):.3f}")

check_double_rate_v3(time_scale=10.0, time_shape=5)
check_double_rate_v3(time_scale=10.0, time_shape=10)
```

```python
import sys
sys.path.insert(0, '../../src')
from workflow import workflow
from bf_simulator import simulator
from priors import prior
import numpy as np
```

```python
true_params = prior()
print("true:", true_params)

fake_data = simulator.sample(1)
fake_data["nu"] = true_params["nu"].reshape(1, -1)
fake_data["alpha1"] = np.array([[true_params["alpha1"]]])
fake_data["alpha2"] = np.array([[true_params["alpha2"]]])
fake_data["tau"] = np.array([[true_params["tau"]]])

posterior_samples = workflow.sample(conditions=fake_data, num_samples=500)

for key in ["nu", "alpha1", "alpha2", "tau"]:
    est = posterior_samples[key].mean(axis=1)
    print(f"{key}: true={true_params[key]}, estimated mean={est}")
```

```python
errors = {"nu": [], "alpha1": [], "alpha2": [], "tau": []}

for _ in range(20):
    true_params = prior()

    fake_data = simulator.sample(1)
    fake_data["nu"] = true_params["nu"].reshape(1, -1)
    fake_data["alpha1"] = np.array([[true_params["alpha1"]]])
    fake_data["alpha2"] = np.array([[true_params["alpha2"]]])
    fake_data["tau"] = np.array([[true_params["tau"]]])

    posterior_samples = workflow.sample(conditions=fake_data, num_samples=500)

    for key in ["nu", "alpha1", "alpha2", "tau"]:
        est = posterior_samples[key].mean(axis=1)[0]
        true_val = np.atleast_1d(true_params[key])
        err = np.abs(est - true_val).mean()
        errors[key].append(err)

for key, vals in errors.items():
    print(f"{key}: mean absolute error = {np.mean(vals):.3f}")
```

```python
prior_means = {"nu": np.array([2.502, 2.502]), "alpha1": 1.005, "alpha2": 16.087, "tau": 0.159}
baseline_errors = {"nu": [], "alpha1": [], "alpha2": [], "tau": []}

for _ in range(20):
    true_params = prior()
    for key in ["nu", "alpha1", "alpha2", "tau"]:
        true_val = np.atleast_1d(true_params[key])
        guess = np.atleast_1d(prior_means[key])
        err = np.abs(guess - true_val).mean()
        baseline_errors[key].append(err)

for key, vals in baseline_errors.items():
    print(f"{key}: naive baseline MAE = {np.mean(vals):.3f}")
```

```python
from simulator import simulate_dataset

errors = {"nu": [], "alpha1": [], "alpha2": [], "tau": []}

for _ in range(20):
    true_params = prior()

    trial_data = simulate_dataset(
        true_params["nu"], true_params["alpha1"], true_params["alpha2"], true_params["tau"]
    )

    fake_data = {
        "n": np.array([[1000]]),
        "nu": true_params["nu"].reshape(1, -1),
        "alpha1": np.array([[true_params["alpha1"]]]),
        "alpha2": np.array([[true_params["alpha2"]]]),
        "tau": np.array([[true_params["tau"]]]),
        "first_response_time": trial_data["first_response_time"].reshape(1, -1),
        "correct": trial_data["correct"].reshape(1, -1),
        "double_flag": trial_data["double_flag"].reshape(1, -1),
        "second_response_time": trial_data["second_response_time"].reshape(1, -1),
    }

    posterior_samples = workflow.sample(conditions=fake_data, num_samples=500)

    for key in ["nu", "alpha1", "alpha2", "tau"]:
        est = posterior_samples[key].mean(axis=1)[0]
        true_val = np.atleast_1d(true_params[key])
        err = np.abs(est - true_val).mean()
        errors[key].append(err)

for key, vals in errors.items():
    print(f"{key}: mean absolute error = {np.mean(vals):.3f}")
```

```python
batch_true = [prior() for _ in range(5)]
batch_data = {"n": np.array([[1000]]*5), "nu": [], "alpha1": [], "alpha2": [], "tau": [],
              "first_response_time": [], "correct": [], "double_flag": [], "second_response_time": []}

for tp in batch_true:
    td = simulate_dataset(tp["nu"], tp["alpha1"], tp["alpha2"], tp["tau"])
    batch_data["nu"].append(tp["nu"])
    batch_data["alpha1"].append(tp["alpha1"])
    batch_data["alpha2"].append(tp["alpha2"])
    batch_data["tau"].append(tp["tau"])
    batch_data["first_response_time"].append(td["first_response_time"])
    batch_data["correct"].append(td["correct"])
    batch_data["double_flag"].append(td["double_flag"])
    batch_data["second_response_time"].append(td["second_response_time"])

for key in ["nu", "first_response_time", "correct", "double_flag", "second_response_time"]:
    batch_data[key] = np.array(batch_data[key])

for key in ["alpha1", "alpha2", "tau"]:
    batch_data[key] = np.array(batch_data[key]).reshape(-1, 1)

posterior_samples = workflow.sample(conditions=batch_data, num_samples=500)
print("worked, shape check:", posterior_samples["nu"].shape)
```
